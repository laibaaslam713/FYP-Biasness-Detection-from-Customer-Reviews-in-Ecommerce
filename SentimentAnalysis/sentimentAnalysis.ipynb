```python
def sentiment_analysis(review_rating):
    print(review_rating)    
    if review_rating == 3:
        sentiment = 'neutral'
    elif review_rating == 1 or review_rating == 2:
        sentiment = 'negative'
    elif review_rating == 4 or review_rating == 5:
        sentiment = 'positive'
    else:
        sentiment = 'unknown'
    return sentiment
```

```python
df['sentiment'] = df['Review Rating'].apply(sentiment_analysis)
df.head()
```
