```python
import json
with open(r'E:\Post ADP\Capstone Project\Scrapping from Amazon\datasets\reviews_dataset\flipkart_kids_furniture.json', 'r') as file:
    total_product_details = json.load(file)  # array for storing complete detail of all products
for i, link in enumerate(links):
    if i > 389 and i < len(links):
        product_price = ''
        product_description = ''
        product_image = ''
        reviewer_ID = ''
        reviewer_helpful_vote = '0'
        reviewer_name = ''
        review_title = ''
        review_text=''
        location=''
        review_date=''
        verified_purchase=''
        review_rating='0'
        review_helpful_vote='0'
        product_categories = ''
        total_rating= 0
        driver.get(link)
        time.sleep(2)
        product_url = driver.current_url
        if product_url:
            Product_ID = product_url.split("pid=")[1].split('&')[0]
        try:
            product_title = driver.find_element(By.CSS_SELECTOR, ".VU-ZEz").text
        except:
            print("title not found")
        try:
            product_description = driver.find_element(By.CSS_SELECTOR, '#container > div > div._39kFie.N3De93.JxFEK3._48O0EI > div.DOjaWF.YJG4Cf > div.DOjaWF.gdgoEp.col-8-12 > div.DOjaWF.gdgoEp > div:nth-child(2) > div > div._4gvKMe > div.yN\+eNk.w9jEaj > p').text
        except:
            print("product description not found")
        try:
            product_categories = driver.find_element(By.CSS_SELECTOR, "#container > div > div._39kFie.N3De93.JxFEK3._48O0EI > div.DOjaWF.YJG4Cf > div.DOjaWF.gdgoEp.col-8-12 > div:nth-child(1) > div:nth-child(1) > div > div:nth-child(3) > a").text
        except:
            print("Product categories not found")
        try:
            click_on_show_more = driver.find_element(By.CSS_SELECTOR, '.QqFHMw._4FgsLt')
            click_on_show_more.click()
            product_features = driver.find_element(By.CSS_SELECTOR, '._1OjC5I').text
        except:
            print("product features not found")
        try:
            product_price = driver.find_element(By.CSS_SELECTOR, ".Nx9bqj.CxhGGd").text
            product_price = product_price.replace("₹", "").strip()
            print(product_price)
        except:
            print("product price not found")
        try:
            product_image = driver.find_element(By.CSS_SELECTOR,".DByuf4.IZexXJ.jLEJ7H")
            product_image = product_image.get_attribute("src")
        except:
            print("product image not found")
        try:
            Brand_name = driver.find_element(By.CSS_SELECTOR, "#sellerName > span > span").text
            # Brand_name = Brand_name.find_element(By.CSS_SELECTOR, '.a-span9').text
        except:
            print("Brand name not found")
        try:
            avg_product_rating = driver.find_element(By.CSS_SELECTOR,".XQDdHH",).text
        except:
            print("average product rating not found")
        try:
            total_rating = driver.find_element(By.XPATH, "/html/body/div[1]/div/div[3]/div[1]/div[2]/div[2]/div/div[2]/div/div/span[2]/span/span[1]",).text
            print(f"total rating: {total_rating}")
            if total_rating:
                total_rating = ''.join(total_rating.split(' ')[0].split(','))
            else:
                print("not found")

            print(f"total rating 1: {total_rating}")
        except:
            print("total rating not found")
        # click on more reviews link for accessing all reviews of a product
        try:
            click_on_more_reviews = driver.find_element(By.CSS_SELECTOR, '._23J90q.RcXBOT')
            click_on_more_reviews.click()
            time.sleep(2)
        except:
            print("more review link not found")
        # found_reviews = 0
        counter = 1
        while True:                                                        
            
            for j in range(4, 13):
                review_helpful_vote = '0'
                try:
                    reviews_div = driver.find_elements(By.CSS_SELECTOR, f'#container > div > div.VLIitu > div > div.DOjaWF.YJG4Cf.col-12-12 > div.DOjaWF.gdgoEp.col-9-12 > div:nth-child({j}) > div > div > div')

                    time.sleep(2)
                except:
                    print("review div not found")
                    break
                try:
                    reviewer_name = driver.find_element(By.CSS_SELECTOR, f"#container > div > div.VLIitu > div > div.DOjaWF.YJG4Cf.col-12-12 > div.DOjaWF.gdgoEp.col-9-12 > div:nth-child({j}) > div > div > div > div.row.gHqwa8 > div.row > p._2NsDsF.AwS1CA").text
                except:                                                    
                    print("Reviewer name not found")
                try:
                    review_title = driver.find_element(By.CSS_SELECTOR, f"#container > div > div.VLIitu > div > div > div.DOjaWF.gdgoEp.col-9-12 > div:nth-child({j}) > div > div > div > div:nth-child(1) > p",).text
                except:
                    print("review_title not found")
                try:
                    review_text = driver.find_element(By.CSS_SELECTOR, f"#container > div > div.VLIitu > div > div.DOjaWF.YJG4Cf.col-12-12 > div.DOjaWF.gdgoEp.col-9-12 > div:nth-child({j}) > div > div > div > div:nth-child(2) > div > div > div",).text
                except:
                    print("review text not found")
                try:
                    review_date = driver.find_element(By.CSS_SELECTOR, f'#container > div > div.VLIitu > div > div.DOjaWF.YJG4Cf.col-12-12 > div.DOjaWF.gdgoEp.col-9-12 > div:nth-child({j}) > div > div > div > div.row.gHqwa8 > div.row > p:nth-child(5)').text
                    # print("review date: ", review_date)
                except:
                    print("date not found")
                
                try:
                    try:
                        verified_purchase_element = driver.find_element(
                            By.XPATH, f"/html/body/div/div/div[3]/div/div[1]/div[2]/div[{j}]/div/div/div/div[3]/div[1]/p[2]"
                        )
                        verified_purchase = verified_purchase_element.text
                        parts = verified_purchase.split(", ")
                        verified_purchase = parts[0]  # "Certified Buyer"
                        location = parts[1] 
                        print(f"Location: {location}")
                        print(f"Verified purchase (path 1): {verified_purchase}")
                    except:
                        # If the first XPath fails, try the second XPath
                        try:
                            verified_purchase_element = driver.find_element(
                                By.XPATH, f"/html/body/div/div/div[3]/div/div[1]/div[2]/div[{j}]/div/div/div/div[4]/div[1]/p[2]"
                            )
                            verified_purchase = verified_purchase_element.text
                            parts = verified_purchase.split(", ")
                            verified_purchase = parts[0]  # "Certified Buyer"
                            location = parts[1] 
                            print(f"Location: {location}")
                            print(f"Verified purchase (path 2): {verified_purchase}")
                        except:
                            # If neither XPath works, print a message
                            print("Verified purchase not found")
                except:
                    print("====verified purchase not found====")
                try:
                    review_rating = driver.find_element(By.CSS_SELECTOR, f"#container > div > div.VLIitu > div > div.DOjaWF.YJG4Cf.col-12-12 > div.DOjaWF.gdgoEp.col-9-12 > div:nth-child({j}) > div > div > div > div:nth-child(1) > div").text
                    # review_rating=review_rating.split[0]
                    print("review_rating: ", review_rating)
                except:
                    print("review rating not found")
                try:
                    review_helpful_vote = driver.find_element(By.CSS_SELECTOR,f"#container > div > div.VLIitu > div > div.DOjaWF.YJG4Cf.col-12-12 > div.DOjaWF.gdgoEp.col-9-12 > div:nth-child({j}) > div > div > div > div.row.gHqwa8 > div.vMTK1V > div > div.qhmk-f > div:nth-child(1) > span",).text 
                    # print("review helpful vote: ", review_helpful_vote)
                except:
                    print("review helpful vote not found")
                scrapedInfo = {
                    "productTitle": product_title if isinstance(product_title, str) else '',
                    "ASIN": Product_ID if isinstance(Product_ID, str) else '',
                    "Price": product_price if isinstance(product_price, str) else '',
                    "Category": product_categories if isinstance(product_categories, str) else '',
                    "Features": product_features if isinstance(product_features, str) else '',
                    "Product Image": product_image if isinstance(product_image, str) else '',
                    "Brand Name": Brand_name if isinstance(Brand_name, str) else '',
                    "Total Rating": int(total_rating) if isinstance(total_rating, str) else 0,
                    "Average Rating": float(avg_product_rating) if isinstance(avg_product_rating, str) else 0.0,
                    "Description": product_description if isinstance(product_description, str) else '',
                    # "Reviewer ID": reviewer_ID if isinstance(reviewer_ID, str) else '',
                    "Reviewer Name": reviewer_name if isinstance(reviewer_name, str) else '',
                    # "Reviewer HelpfulVote": reviewer_helpful_vote,
                    "Review Title": review_title if isinstance(review_title, str) else '',
                    "Review Text": review_text if isinstance(review_text, str) else '',
                    "Location": location if isinstance(location, str) else '',
                    "Review Date": review_date if isinstance(review_date, str) else '',
                    "Verified Purchase": True if verified_purchase == "Certified Buyer" else False,
                    "Review Rating": float(review_rating) if isinstance(review_rating, str) else 0.0,
                    "Review helpful vote": int(review_helpful_vote)
                }
                # print(scrapedInfo)
                total_product_details.append(scrapedInfo)
                print('after creating object')
            try:
                # Wait until the "Next" anchor tag is available
                next_anchor = WebDriverWait(driver, 10).until(
                    EC.presence_of_element_located((By.XPATH, "//a[@class='_9QVEpD' and span='Next']"))
                )
                
                # Get the 'href' attribute of the anchor tag
                next_href = next_anchor.get_attribute('href')
                print(f"Next page link: {next_href}")
                
                # Optionally navigate to the next page
                driver.get(next_href)
                time.sleep(2)

            except:
                print(f"Error locating the 'Next' anchor tag")
                break
        print("=====================")
        with open(r'E:\Post ADP\Capstone Project\Scrapping from Amazon\datasets\reviews_dataset\flipkart_kids_furniture.json', 'w') as json_file:
            json.dump(total_product_details, json_file, indent=4)
        print(len(total_product_details))
        print("=====================")
        for j in range(counter):
            driver.back()
            time.sleep(2)
        # driver.back()
        # time.sleep(2)
        print(f"{i} ====== links done =========")
    
    
```
