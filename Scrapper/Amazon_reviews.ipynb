```python
import json
with open('Snack_Food.json', 'r') as file:
    total_product_details = json.load(file)  # array for storing complete detail of all products
for i, link in enumerate(links):
    if i > 66 and i < len(links):
        product_price = ''
        product_description = ''
        product_image = ''
        reviewer_ID = ''
        reviewer_helpful_vote = '0'
        reviewer_name = ''
        review_title = ''
        review_text=''
        location=''
        review_date=''
        verified_purchase=''
        review_rating='0'
        review_helpful_vote='0'
        driver.get(link)
        time.sleep(2)
        product_url = driver.current_url
        if product_url:
            Product_ID = product_url.split("dp/")[1].split('/')[0]
        try:
            product_title = driver.find_element(By.CSS_SELECTOR, ".a-size-large.product-title-word-break").text
        except:
            print("title not found")
        try:
            product_description = driver.find_element(By.CSS_SELECTOR, '.a-unordered-list.a-vertical.a-spacing-mini').text
        except:
            print("product description not found")
        try:
            product_categories = driver.find_element(By.CSS_SELECTOR, ".a-unordered-list.a-horizontal.a-size-small").text
        except:
            print("Product categories not found")
        try:
            product_features = driver.find_element(By.CSS_SELECTOR, '.a-section.a-spacing-small.a-spacing-top-small').text
        except:
            print("product features not found")
        try:
            product_price = driver.find_element(By.CSS_SELECTOR, ".a-price-whole").text
        except:
            print("product price not found")
        try:
            product_image = driver.find_element(By.CSS_SELECTOR,"#landingImage")
            product_image = product_image.get_attribute("src")
        except:
            print("product image not found")
        try:
            Brand_name = driver.find_element(By.CSS_SELECTOR, ".a-spacing-small.po-brand")
            Brand_name = Brand_name.find_element(By.CSS_SELECTOR, '.a-span9').text
        except:
            print("Brand name not found")
        try:
            avg_product_rating = driver.find_element(By.CSS_SELECTOR,".a-size-base.a-color-base",).text
        except:
            print("average product rating not found")
        try:
            total_rating = driver.find_element(By.CSS_SELECTOR, ".a-size-base.a-color-secondary",).text
            print(f"total rating: {total_rating}")
            if total_rating[0] == '(':
                total_rating = ''.join(total_rating.split('(')[1].split(')')[0].split(','))
            else:
                total_rating = ''.join(total_rating.split(' ')[0].split(','))

            print(f"total rating 1: {total_rating}")
        except:
            print("total rating not found")
        # click on more reviews link for accessing all reviews of a product
        try:
            click_on_more_reviews = driver.find_element(
                By.CSS_SELECTOR, ".a-link-emphasis.a-text-bold"
            )
            more_reviews_link = click_on_more_reviews.get_attribute("href")
            driver.get(more_reviews_link)
            time.sleep(2)
        except:
            print("more review link not found")
            break
        # found_reviews = 0
        counter = 1
        while True:
            
            for j in range(2, 10):
                review_helpful_vote = '0'
                try:
                    click_on_review = driver.find_element(
                        By.XPATH,
                        f"/html/body/div[1]/div[2]/div/div[1]/div/div[1]/div[5]/div[3]/div/div[{j}]/div/div/div[2]/a/span[2]",
                    )
                    
                    click_on_review.click()
                    # found_reviews = found_reviews + 1
                    # print(found_reviews)
                    time.sleep(2)
                    # print("review found")
                except:
                    print("review link not found")
                    break
                # review_div = driver.find_element(By.CSS_SELECTOR, ".a-fixed-right-grid-col.a-col-left")
                # try:
                try:
                    Brand_name = driver.find_element(By.XPATH, "/html/body/div[1]/div[1]/div[1]/div/div/div[2]/span[1]/div/div/div[2]/div[2]/div/span/a").text
                    # print(Brand_name)
                    # Brand_name = Brand_name.find_element(By.CSS_SELECTOR, '.a-span9').text
                except:
                    print("Brand name not found")
                try:
                    reviewer_name = driver.find_element(By.CSS_SELECTOR, ".a-profile-name").text
                except:
                    print("Reviewer name not found")
                try:
                    review_title = driver.find_element(By.XPATH, "/html/body/div[1]/div[1]/div[1]/div/div/div[1]/div/div/div/div/div[2]/a/span[2]",).text
                except:
                    print("review_title not found")
                try:
                    review_text = driver.find_element(By.XPATH, "/html/body/div[1]/div[1]/div[1]/div/div/div[1]/div/div/div/div/div[4]/span/span",).text
                except:
                    print("review text not found")
                try:
                    review_date_text = driver.find_element(By.CSS_SELECTOR, '.a-size-base.a-color-secondary.review-date').text
                    # Example text: "Reviewed in India on 3 October 2024"
                    
                    # Splitting based on "on" to separate location and date
                    location_date_split = review_date_text.split(" on ")
                    
                    # Get the location and date separately
                    location = location_date_split[0].replace("Reviewed in ", "").strip()  # Removes "Reviewed in" prefix
                    review_date = location_date_split[1].strip()  # Gets the date part
                except:
                    print("date and location not found")
                
                try:
                    verified_purchase = driver.find_element(By.CSS_SELECTOR, ".a-size-mini.a-color-state.a-text-bold").text
                except:
                    print("verified purchase not found")
                try:
                    review_rating_anchor = driver.find_element(By.CSS_SELECTOR, "a.a-size-base.a-link-normal.review-title.a-color-base.review-title-content.a-text-bold")
                    review_rating = review_rating_anchor.find_element(By.TAG_NAME, "i")
                    review_rating = review_rating.get_attribute('class')
                    # a-icon a-icon-star a-star-5 review-rating
                    review_rating = review_rating.split(' ')[2]
                    review_rating = review_rating.split('-')[2]
                except:
                    print("review rating not found")
                try:
                    review_helpful_vote = (
                        driver.find_element(By.CSS_SELECTOR,".a-size-base.a-color-tertiary.cr-vote-text",).text 
                        if driver.find_elements(
                            By.CSS_SELECTOR,
                            ".a-size-base.a-color-tertiary.cr-vote-text",
                        )
                        else '0'
                    )
                    review_helpful_vote = review_helpful_vote.split(' ')[0] if review_helpful_vote else '0'
                    print(f"helpful votes: {''.join(review_helpful_vote.split(','))}")
                    if review_helpful_vote == "One":
                        review_helpful_vote = '1'
                # else:
                    # review_helpful_vote = ''.join(reviewer_helpful_vote.split(','))
                except:
                    print("review helpful vote not found")
                try:
                    reviewer_profile_link = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.CSS_SELECTOR, '.a-profile')))
                    reviewer_profile_link.click()
                    time.sleep(2)
                    profile_url = driver.current_url
                except:
                    print("reviewer profile link not found")
                
                try:
                    reviewer_helpful_vote = driver.find_element(By.XPATH, f'/html/body/div[1]/div[1]/div/div/div[4]/div[1]/div[2]/div/div/span[1]').text
                    reviewer_helpful_vote = ''.join(reviewer_helpful_vote.split(','))
                    reviewer_ID = profile_url.split("account.")[1].split('/')[0]
                    driver.back()
                    print('before creating object')
                except:
                    print("reviewer id and reviewer helpful vote not found")
                # if review_helpful_vote:
                #     scrapedInfo["Review helpful vote"] = review_helpful_vote
                # except AttributeError:
                # print("======/======/=======")
                # print("======/======/=======")
                driver.back()
                time.sleep(2)
                # except:
                #     driver.back()
                #     time.sleep(2)
                #     print(f"no more review found on page {counter}")
                #     # break
                # # driver.back()
                scrapedInfo = {
                    "productTitle": product_title if isinstance(product_title, str) else '',
                    "ASIN": Product_ID if isinstance(Product_ID, str) else '',
                    "Price": product_price if isinstance(product_price, str) else '',
                    "Category": product_categories if isinstance(product_categories, str) else '',
                    "Features": product_features if isinstance(product_features, str) else '',
                    "Product Image": product_image if isinstance(product_image, str) else '',
                    "Brand Name": Brand_name if isinstance(Brand_name, str) else '',
                    "Total Rating": int(total_rating) if isinstance(total_rating, str) else 0,
                    "Average Rating": float(avg_product_rating) if isinstance(avg_product_rating, str) else 0.0,
                    "Description": product_description if isinstance(product_description, str) else '',
                    "Reviewer ID": reviewer_ID if isinstance(reviewer_ID, str) else '',
                    "Reviewer Name": reviewer_name if isinstance(reviewer_name, str) else '',
                    "Reviewer HelpfulVote": reviewer_helpful_vote,
                    "Review Title": review_title if isinstance(review_title, str) else '',
                    "Review Text": review_text if isinstance(review_text, str) else '',
                    "Location": location if isinstance(location, str) else '',
                    "Review Date": review_date if isinstance(review_date, str) else '',
                    "Verified Purchase": True if verified_purchase == "Verified Purchase" else False,
                    "Review Rating": float(review_rating) if isinstance(review_rating, str) else 0.0,
                    "Review helpful vote": int(review_helpful_vote)
                }
                # print(scrapedInfo)
                total_product_details.append(scrapedInfo)
                print('after creating object')
            try:
                next_review_page = driver.find_element(
                    By.CSS_SELECTOR, ".a-last a"
                )
                # print("next page found")
                if next_review_page:
                    next_review_page.click()
                    time.sleep(2)
                    counter += 1
                else:
                    driver.back()
                    time.sleep(2)
            except:
                print("next page not found")
                break
        print("=====================")
        with open('Snack_Food.json', 'w') as json_file:
            json.dump(total_product_details, json_file, indent=4)
        print(len(total_product_details))
        print("=====================")
        for j in range(counter):
            driver.back()
            time.sleep(2)
        # driver.back()
        # time.sleep(2)
        print(f"{i} ====== links done =========")
    
```
